```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df1=pd.read_excel('/content/Unemployment in India.xlsx')
print(df1)
```

```python
df1.info()
```

```python
df2=pd.read_excel('/content/Unemployment_Rate_upto_11_2020.xlsx')
print(df2)
```

```python
df2.info()
```

```python
print(df1.isnull().sum())
```

```python
print(df2.isnull().sum())
```

```python
df1=df1.dropna()
df2=df2.dropna()
```

```python
print(df1.describe())
print(df2.describe())
```

# **Regional Analysis On Unemployment in India Dataset**

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Area',y='Estimated Unemployment Rate (%)',data=df1)
plt.xticks(rotation=45)
plt.title('Unemployment Rate by Area')
plt.show()
```

# **Regional Analysis On Unemployment Rate Upto 11_2020 Dataset**




```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Region',y='Estimated Unemployment Rate (%)',data=df2)
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.title('Unemployment Rate by Region')
plt.show()
```

# **Correlation Analysis On Unemployment In India Dataset**

```python
correlation_matrix=df1[['Estimated Unemployment Rate (%)','Estimated Employed','Estimated Labour Participation Rate (%)']].corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# **Correlation Analysis On Unemployment Rate Upto 11_2020 Dataset**

```python
correlation_matrix=df2[['Estimated Unemployment Rate (%)','Estimated Employed','Estimated Labour Participation Rate (%)']].corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# **Grouping by 'Area' and calculate mean unemployment rate**

```python
grouped_df1=df1.groupby('Area')['Estimated Unemployment Rate (%)'].mean()
print(grouped_df1)
```

# **Group by 'Region' and calculate mean unemployment rate**

```python
grouped_df2=df2.groupby('Region')['Estimated Unemployment Rate (%)'].mean()
print(grouped_df2)
```

# **Filtering areas with unemployment rate greater than 10% from Unemployment In India Dataset**

```python
high_unemployment_areas=df1[df1['Estimated Unemployment Rate (%)'] > 20]
sorted= high_unemployment_areas.sort_values('Estimated Unemployment Rate (%)')
display(sorted)
```

# **Filtering areas with unemployment rate greater than 10% from Unemployment Rate Upto 11_2020 Dataset**

```python
high_unemployment_areas = df2[df2['Estimated Unemployment Rate (%)'] > 20]
sorted = high_unemployment_areas.sort_values('Estimated Unemployment Rate (%)')
display(sorted)
```
